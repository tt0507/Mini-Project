```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('data/data.csv')
```

Drop unnecessary columns and rows. We drop the 2019 year data as there are only 405 non-null values.

```python
# drop last five rows
all_data = all_data.iloc[:-5, :]

# drop columns
all_data = all_data.drop(columns=['Series Code', 'Country Name', '2019 [YR2019]'])

all_data.head()
```

Since the type of cell values are a string, the data value is turned into a float. For cell values that have no value, it was replaced is `NaN`.

```python
for column in all_data.columns[2:]:
    for row in range(len(all_data)):
        if all_data.iloc[row][column] == '..':
            all_data.iloc[row][column] = np.nan
        else:
            all_data.iloc[row][column] = float(all_data.iloc[row][column])
```

```python
data_cell = all_data.loc[:, '2000 [YR2000]':].T.fillna(all_data.loc[:, '2000 [YR2000]':].mean(axis=1)).T
```

```python
new_columns = [str(x) for x in range(2000, 2019, 1)]
data_cell = data_cell.rename(columns = dict(zip(data_cell.columns, new_columns)))
```

```python
data_cell['series'] = all_data['Series Name']
data_cell['country'] = all_data['Country Code']
data_cell = data_cell.reindex(columns=['series', 'country', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017',
       '2018'])
```
